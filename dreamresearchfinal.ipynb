```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
).to("cuda")

pipe.enable_xformers_memory_efficient_attention()
pipe.safety_checker = None

print("✅ Model loaded")
```

```python
from peft import TaskType
print(TaskType.__members__)
```

```python
from diffusers import AutoencoderKL, UNet2DConditionModel
from transformers import CLIPTokenizer
from peft import get_peft_model, LoraConfig

vae = pipe.vae
tokenizer = pipe.tokenizer
text_encoder = pipe.text_encoder
unet = pipe.unet

lora_config = LoraConfig(
    r=4,
    lora_alpha=16,
    target_modules=["to_q", "to_v"],
    lora_dropout=0.0,
    bias="none",
)
unet = get_peft_model(unet, lora_config)
```

```python
from PIL import Image
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
import os
class ImageDataset(Dataset):
    def __init__(self, folder, prompt):
        self.prompt = prompt
        self.images = [os.path.join(folder, f) for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = Image.open(self.images[idx]).convert("RGB").resize((512, 512))
        image = np.array(image)
        image = torch.tensor(image).permute(2, 0, 1).float() / 255.0
        return {"pixel_values": image, "prompt": self.prompt}

instance_prompt = "a painting in kathy style"
dataset = ImageDataset("data/blended_art/Test", instance_prompt)
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
import torch, gc
gc.collect()
torch.cuda.empty_cache()
```

```python
import torch
from torch.utils.data import DataLoader
from torch.optim import Adam, AdamW
from torch.utils.checkpoint import checkpoint

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

optimizer = AdamW(unet.parameters(), lr=1e-6)
unet.train()

# Set gradient accumulation steps and batch size
accumulation_steps = 16
batch_size = 1  # You can further reduce the batch size if necessary

# Dataloader with batch size
dataloader = DataLoader(dataset, batch_size=batch_size, pin_memory=True, shuffle=True)

# Move the model to full precision for debugging
unet = unet.to(device).float()  # Ensure the model is in full precision for debugging

# Function for checkpointing UNet forward pass
def checkpointed_unet(input_image, timesteps, encoder_hidden_states):
    def custom_forward(*inputs):
        return unet(*inputs).sample  # Ensure correct output from UNet
    return checkpoint(custom_forward, input_image, timesteps, encoder_hidden_states)

# Gradient clipping to prevent exploding gradients
def clip_gradients(model, max_norm=1.0):
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm)

# Function to clamp predictions and loss values
def clamp_tensor(tensor, min_val=-1.0, max_val=1.0):
    return torch.clamp(tensor, min=min_val, max=max_val)

for epoch in range(1):  # Training loop, here it's set to 1 epoch
    optimizer.zero_grad()  # Clear previous gradients

    for step, batch in enumerate(dataloader):
        # Move data to GPU and convert to full precision
        images = batch["pixel_values"].to(device).float()  # Move image data to GPU

        # Check for NaNs in input images
        if torch.isnan(images).any() or torch.isinf(images).any():
            print(f"NaN or Inf detected in images at step {step}")
            break

        # Debugging: Check min/max values of images
        print(f"Min/Max values in images: {images.min().item()}, {images.max().item()}")

        # Keep images on the GPU and add the alpha channel there
        alpha_channel = torch.ones(images.shape[0], 1, images.shape[2], images.shape[3], device=images.device).float()
        images = torch.cat((images, alpha_channel), dim=1)

        # Create noise and noisy images
        noise = torch.randn_like(images, requires_grad=True).float() * 0.02  # Reduce noise scale

        # Check for NaNs in noise
        if torch.isnan(noise).any() or torch.isinf(noise).any():
            print(f"NaN or Inf detected in noise at step {step}")
            break

        # Debugging: Check min/max values of noise
        print(f"Min/Max values in noise: {noise.min().item()}, {noise.max().item()}")

        # Reduce the scale of the noise even further (0.02)
        noisy_images = images + noise * 0.02  # Reduced noise scale

        # Check for NaNs in noisy images
        if torch.isnan(noisy_images).any() or torch.isinf(noisy_images).any():
            print(f"NaN or Inf detected in noisy_images at step {step}")
            break

        # Debugging: Check min/max values of noisy images
        print(f"Min/Max values in noisy_images: {noisy_images.min().item()}, {noisy_images.max().item()}")

        timesteps = torch.randint(0, 500, (images.shape[0],), device=device).long()

        # Tokenize and process text data
        tokenizer_output = tokenizer(batch["prompt"], return_tensors="pt", padding="max_length", truncation=True, max_length=77)
        input_ids = tokenizer_output["input_ids"].to(device)  # Access the correct key for input_ids
        encoder_hidden_states = text_encoder(input_ids)[0].float()  # Set to full precision

        # Forward pass with checkpointing
        noise_pred = checkpointed_unet(noisy_images, timesteps, encoder_hidden_states)

        # Check for NaNs in the prediction
        if torch.isnan(noise_pred).any() or torch.isinf(noise_pred).any():
            print(f"NaN or Inf detected in noise_pred at step {step}")
            print(f"Noise_pred values: {noise_pred}")
            break

        # Clamp the predictions to avoid NaN/Inf
        noise_pred = clamp_tensor(noise_pred)

        # Ensure the loss is computed with float32 precision for stability
        noise_pred = noise_pred.to(torch.float32)
        noise = noise.to(torch.float32)  # Ensure noise is also in float32 for loss calculation

        loss = torch.nn.functional.mse_loss(noise_pred, noise)

        if torch.isnan(loss).any() or torch.isinf(loss).any():
            print(f"NaN or Inf detected in loss at step {step}")
            break

        loss.backward()

        for name, param in unet.named_parameters():
            if torch.isnan(param.grad).any() or torch.isinf(param.grad).any():
                print(f"NaN or Inf detected in gradients of {name}")
                break

        clip_gradients(unet)

        if (step + 1) % accumulation_steps == 0:
            optimizer.step()  # Update the weights
            optimizer.zero_grad()  # Clear gradients for the next step
        if step % 10 == 0:
            print(f"Step {step}, Loss: {loss.item()}")

        torch.cuda.empty_cache()

print("✅ Training done.")
```
